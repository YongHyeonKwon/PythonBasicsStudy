```python
def cal(num1,num2,op):
    #     num1, num2, op
    if(op == '+'):
        # + 
        return num1+num2
    elif(op == '-'):
        #  
        return num1-num2
```

```python
num1 = int(input('    >> '))
num2 = int(input('    >> '))
op = input(" (+,-) >> ")
result = cal(num1, num2, op)
print(" : {}".format(result))
# print(f' : {result}')
```

```python
def cal(num1,num2,op):
    """    
    num1: int, num2: int, op: string"""
    if(op == '+'):
        # + 
        return num1+num2
    elif(op == '-'):
        #  
        return num1-num2
```

```python
cal()
```

```python
#  return  
# 1. return  
def add_sub(num1, num2):
    return num1+num2, num1-num2, num1-num2 # packing : 
```

```python
a,b,c = add_sub(10,7)
print(a)
print(b)
print(c)
# a,b,c       -> unpacking :  
```

```python
#   
def add(*num):
    total = 0
    for i in num:
        total += i # 1,2,3,4,5
    return total
    
add(1,2,3,4,8,9,51,1231)
```
