### 문자열 접근하는 방법
- 문자열 -> 데이터
- 인덱싱, 슬라이싱

```python
# 인덱싱 -> 한자리만 가리킴
s = "My name is JY"
print(s[0])
print(s[8])
print(s[-2])
print(s[-1])
```

```python
# 슬라이싱 -> 특정 문자열 접근
# [시작인덱스:끝인덱스+1]
# My 슬라이싱
s = "My name is JY"
print(s[0:2])

# name
s[3:7]
# is
print(s[8:10:1])
```

```python
# 생략할 수 있는 인덱스 번호
# s 변수에서 맨 첫 번째 인덱스부터 name e까지 슬라이싱

s = "My name is JY"
print(s[0:7])
print(s[:7]) # 시작값의 기본값:0 (생략가능)

print(s[8:13])
print(s[8:]) # 끝값의 기본값: 문장의 끝 (생략 가능)
print(s[:]) # 모든 문자열에 접근 1칸씩 뛰면서 접근
print(s[::2]) # 모든 문자열에 접근 2칸씩 뛰면서 접근
```

```python
day = "2020년 3월 3일의 날씨는 맑음입니다."
print('날짜:', day[0:11])
print('날씨:' , day[17:19])
```

```python
# 문자열 관련된 함수 
st = day.find('2020') # 해당 단어가 어디부터 시작되고 있는지 출력 0
ed = day.find('3일') # 9 + 2
print(st, ed)
day[st:ed+2]
```

```python
# 문자열 포매팅
# 문자열 안의 특정한 값을 바꿔야 할 때 사용
now = '오늘은 7월26일입니다.'
# now[:2] = '내일' # 에러 발생 - 슬라이싱해서 수정 불가능
```

```python
# 일자를 변경해보기 26 -> 27
mth = 7
day = 28
'오늘은 {}월 {}일입니다.'.format(mth,day)
```

```python
# f문자열 포매팅
hour = 6
mnt = 45
# '오늘 나는 {}시{}분에 일어났다.'.format(hour,mnt)
f'오늘 나는 {hour}시 {mnt}분에 일어났다.'
```

```python
from datetime import datetime
current_time = datetime.now()
h = current_time.hour
m = current_time.minute
s = current_time.second
f'실행 당시 현재 시간은 {h}시 {m}분 {s}초입니다.'
```

#### 문자열 관련 함수 정리

```python
# 길이 확인하는 함수
len(now)
now[9:len(now)]
```

```python
# 문자열 대소문자 변환 관련 함수
# 소문자를 대문자로 바꾸기 
eng_str = 'hi, My name is JY'
eng_str2 = eng_str.upper() # 해당 변수에 초기화 x
eng_str2
# 대문자를 소문자로 바꾸기 
eng_str3 = eng_str.lower()
eng_str3
```

```python
# '찾을 문자열'이 몇 개 있는지 확인하는 함수
# 'm' 카운트
# 대소 문자 구별
eng_str.lower().count('m')
# .count('M')
```

```python
# 특정 문자열 찾아서 인덱스 번호 확인 (반환)하는 함수
# find()
# index()
eng_str.find('i') # 제일 먼저 찾아지는 'i' 인덱스 값 출력
eng_str.index('i') # 제일 먼저 찾아지는 'i' 인덱스 값 출력

#문자열에 없는 단어 찾아달라고 하기
eng_str.find('z') # 제일 먼저 찾아지는 'i' 인덱스 값 출력
# find는 해당 단어가 없으면 -1 값 반환, 에러 x
eng_str.index('z') # 제일 먼저 찾아지는 'i' 인덱스 값 출력
# index는 해당 단어가 없으면 에러 발생 
```

```python
 # strip() 문자열의 양끝의 공백을 제거하는 함수
'      배고프다!        '.strip()
```

```python
# 문자열1을 문자열2로 변경하는 함수
# replace(문자열1, 문자열2)
eng_str.replace('JY','JH')
```
