```python
# 네임스페이스 (namespace, 이름 공간)
# 프로그래밍 언어에서 특정한 객체(object)를 이름(name)에 따라 구분할 수 있는 범위 
# 전역 네임스페이스, 지역 네임스페이스 
x = 9
print('함수 밖', x)
def temp():
    x = 10 #  함수 안에 있는 x변수 
    print('함수 안:', x)
temp()
```

```python
# global 키워드
x = 9
print('전역 변수', x)
def temp():
    global x # x를 전역 변수로 만듬
    x = 10 #  함수 안에 있는 x변수 
    print('전역 변수 개념으로 사용:', x)
temp()
print('전역 변수2:',x)
```

```python
# 함수 안에 내용 짜보기 : 어떤 수의 약수를 구하는 코드
# 함수로 정의
def divisor(num):
    for i in range(1,num+1):
        if num % i == 0:
            print(i, end = ' ')
            
divisor(100)
```

```python
def cal(num1,num2,op):
    # 함수 안에서 사용할 매개변수명 num1, num2, op
    if(op == '+'):
        # + 연산
        return num1+num2
    elif(op == '-'):
        # 빼기 연산
        return num1-num2
```

```python
num1 = int(input('첫 번째 정수 입력 >> '))
num2 = int(input('두 번째 정수 입력 >> '))
op = input("연산자 입력(+,-) >> ")
result = cal(num1, num2, op)
print("결과 : {}".format(result))
# print(f'결과 : {result}')
```

```python
def cal(num1,num2,op):
    """덧셈과 뺄셈을 계산하는 함수 
    num1: int, num2: int, op: 연산자string"""
    if(op == '+'):
        # + 연산
        return num1+num2
    elif(op == '-'):
        # 빼기 연산
        return num1-num2
```

```python
cal()
```

```python
def greeting(name):
    """이름이 입력되면 인사를 하는 기능
    name type : str 
    return type : str"""
    return print(f"'{name} 안녕 ~ '")
```

```python
greeting('지영')
```

```python
# 타입 힌팅(hinting) : 설명문
def greeting(name: str) -> str:
    return f'{name} 안녕!'
```

```python
greeting.__annotations__#__언더바 두개, 던더 (더블언더바)
```

```python
greeting()
```

```python
# 함수 return 키워드 정리
# 1. return 키워드 정리
def add_sub(num1, num2):
    return num1+num2, num1-num2, num1-num2 # packing : 포장
```

```python
a,b,c = add_sub(10,7)
print(a)
print(b)
print(c)
# a,b,c 로 각각 따로 담을 수 있음 -> unpacking : 포장을 품
```

```python
# 2. 함수 안에서 return을 만나면 함수가 끝남
def temp2():
    for i in range(3): # 3번 반복하는 for
        print(i)# 0
    return i
temp2() # return 맨 마지막 i인 2만 반환
```

```python
# 지수 연산하는 함수 정의
def power_num(num , power = 2):
    return num ** power
```

```python
# 가변 매개 변수
def add(*num):
    total = 0
    for i in num:
        total += i # 1,2,3,4,5
    return total
    
add(1,2,3,4,8,9,51,1231)
```

```python
# 함수 4가지 모형
# 1. 매개변수x, 리턴x
# 2. 매개변수o, 리턴x
# 3. 매개변수x, 리턴o
# 4. 매개변수o, 리턴o
```

```python
# 1. 매개변수x, 리턴x
def hi():
    print('안녕~')
hi()
```

```python
# 4. 매개변수o, 리턴o
def hi4(name):
    return f'{name} 안녕~'

result = hi4('지영')
print(result)
```
