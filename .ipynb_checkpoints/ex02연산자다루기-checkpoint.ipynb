```python
# a > b : a b 
# a >= b : a b  
# a < b : b a 
# a <= b
# a <= b : b a  
# a==b: a b 
# a != b : a b  ( )
```

```python
# random ()
import random 
a = random.randint(1,10)
b = random.randint(1,10)
print(a < b)
print(a > b)
print(a == b)
print(a != b)
print(a, b)
```

```python
print(True and False)
print(True or False)
print(not False) # if    
```

```python
print(2>1 and 2==1)
print(2==2 or 2<1)
print(not 2<1) # if    
```
