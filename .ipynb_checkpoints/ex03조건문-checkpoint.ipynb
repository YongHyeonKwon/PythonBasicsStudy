```python
# 조건문 if
if 2>1: 
    print('해당 조건식은 트루입니다~')
    
if 2<1: # False 조건문 안을 확인 x 셀 실행 멈춤
    print('조건문 안이에요 ~ ')
```

```python
num = int(input("정수 입력 >> "))
if num%3 == 0 and num%5 ==0 :
    print('3과 5의 배수입니다.')
else :
    print('3과 5의 배수가 아닙니다.')
```

```python
num1 = int(input('첫 번째 정수 입력 >>'))
num2 = int(input('두 번째 정수 입력 >>'))
if num1 > num2 :
    print("첫 번째 정수가 더 큽니다")
elif num2 > num1 :
    print('두 번째 정수가 더 큽니다')
else :
    print('두 수가 똑같습니다.')
```

```python
def print_menu():
    print("""
================================================
================---- 자판기 ----================
================================================
== 1.콜라 == 2.사이다 ==  3.물  =======insert===
==--------==----------==--------=======------===
==   600  ==    800   ==  1000  ======= O00원===
=======================================------===
================================================
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==+++++++++++++++     SMHRD    +++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
================================================
================================================
  ===                                      ===
  ===                                      ===""")

print_menu()
money = int(input('금액 넣기 >> '))
menu = int(input('메뉴 선택(번호입력)>>'))

if menu ==1 and money>=600:
    money = money - 600
elif menu == 2 and money>=800:
    money = money - 800
elif menu == 3 and money>=1000:
    money = money - 1000
    
if money < 0:
    print('금액이 부족합니다.')
else: 
    chun = money//1000
    obek = money%1000//500
    bek = money%1000%500//100
    print(f'거스름돈 >> 1000원 {chun}개 500원 {obek}개 100원 {bek}개 ')
```
