```python
# 리스트 선언하기
a = [] # 빈 리스트
b = [1,2,3] # 숫자 데이터를 담은 리스트
c = ['My','name','is','JY'] # 문자열 데이터를 담은 리스트
d = [1,2,'My','name'] # 숫자, 문자열 데이터를 담은 리스트
e = [1,2,['My','name',]] # 이중 리스트 가능
print(a)
print(b)
print(c)
print(d)
print(type(e))
```

#### 리스트 데이터 접근
- 인덱스 번호 활용
- 문자열 인덱싱, 슬라이싱 활용과 동일
- 인덱싱: 무엇인가를 '가리킴'
- 슬라이싱: 무엇인가를 '잘라냄'(범위)
- 대상[인덱스 번호], 리스트[시작:끝+1]

```python
list1 = [2,5,7,9,10]
# 2 접근
list1[0]
# 9 접근
list1[3]
```

```python
list2 = [1,2,3,['a','b','c']]
temp = (list2[3])
print(temp)
print(list2[3][1])
```

```python
# 슬라이싱
# 리스트[시작:끝+1]
list3 = ['지수','문규','다인','수양','성문']
list3[0:2]
# 슬라이싱 결과는 여러 개에 대해 리스트로 묶어서 반환 
```

```python
# '다인','성문' 슬라이싱 해보기
list3[2:5:2]
```

```python
# 맨 뒤에부터 앞에까지 차례대로 출력하기 
list3[::-1]
```

```python
# 리스트는 가변적이다!
# 리스트 더하기
list4 = ['인우','지웅','강련','서준희']
print(list3+list4)

# 리스트 곱하기
list5 = [1,2,3]
print(list5 * 3) # 수학 개념의 연산x, 자료의 크기를 그만큼 늘림 o
```

```python
# 리스트 맨 뒤쪽 값 추가
# 리스트 append(값)
list5.append(4)
list5
```

```python
lst = ['딸기','바나나','수박','체리','포도']
# list.index(값) -> 인덱스 번호 반환하는 함수

str1 = str(input('검색할 문자를 입력하세요 >> '))


if(str1 in lst):
    num = lst.index(str1)
    print(f'{str1}는 리스트 {num}번째 인덱스에 들어있습니다.')
else:
    print(f'{str1}는 리스트에 들어있지 않습니다.')
```

```python
# 문자열대상.count(값)
# 문자열 변수 선언
s = "Hi, My name is Mihee"
s = s.lower()
# 검색 문자 입력 받기
search = input('검색할 문자를 입력하세요 : ')
cnt = 0
# 조건문
if search in s:
    cnt = s.count(search)
    print(f"'{search}'는 {cnt}번 들어 있어요.")
else:
    print(f"'{search}'는 들어있지 않습니다.")
```
