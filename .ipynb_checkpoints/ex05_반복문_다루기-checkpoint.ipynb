```python
#     
import random
com_num = random.randint(1,50) #  

#    break

while True:
    num = int(input('  >> '))
    
    if (com_num < num):
        print(f'{num}   .')
    elif (com_num > num):
        print(f'{num}   .')
    elif (com_num == num):
        print(' .')
        break
```
