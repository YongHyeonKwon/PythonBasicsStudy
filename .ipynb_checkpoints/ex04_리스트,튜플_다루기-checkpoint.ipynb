```python
lst = ['','','','','']
# list.index() ->    

str1 = str(input('   >> '))


if(str1 in lst):
    num = lst.index(str1)
    print(f'{str1}  {num}  .')
else:
    print(f'{str1}   .')
```

```python
# .count()
#   
s = "Hi, My name is Mihee"
s = s.lower()
#    
search = input('   : ')
cnt = 0
# 
if search in s:
    cnt = s.count(search)
    print(f"'{search}' {cnt}  .")
else:
    print(f"'{search}'  .")
```
